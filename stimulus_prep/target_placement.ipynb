# Target Placement 

The target is renderered atop the target object on the vertical midline, and the pixel location specified in the filename

```python
import pandas as pd 
import json
from PIL import Image, ImageDraw
import os
import numpy as np
```

To derive the actual y coordinate, the formula is:


(y_coord/840) * height of the image

```python
# INSERT FOLDER DESTINATION HERE 
dest = ''
```

```python
for img in os.listdir(dest):
    if 'png' in img:
        path = dest + '/' + img 
        image = Image.open(path)  # Replace with the path to your image

        # Convert the image to RGB mode
        image = image.convert("RGB")

        # Get the dimensions (width and height) of the image
        width, height = image.size

        # Create a drawing object
        draw = ImageDraw.Draw(image)

        logged_y_coord = int(img.split('_')[-1][:-4])
        y_coord = height - ((logged_y_coord/840)*height)

        # Define the center and radius of the circle
        center_x, center_y = width//2, y_coord  # Replace with your desired coordinates
        radius = 7  # Replace with your desired radius

        # Define the circle's bounding box
        left = center_x - radius
        top = center_y - radius
        right = center_x + radius
        bottom = center_y + radius

        # Draw the red circle on the image
        draw.ellipse([(left, top), (right, bottom)], fill="red", outline="red", width=4)


        # Resize the image
        resized_image = image.resize((730, 530), Image.LANCZOS)


        # Save the resized image with the specified DPI as a high-quality JPEG
        outputFolder_path = dest + "/"  # Replace with your desired output path

        resized_image.save(outputFolder_path + img, quality=95)  # Adjust quality as needed

```
